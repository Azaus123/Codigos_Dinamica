```python
#hallar magnitud minima velocidad para dejar el techo
#Tambien hallar la distancia D a la que llega
#Esto es solo velocidad horizontal
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.animation import FuncAnimation
from sympy import symbols, Matrix, pi,deg, cos, sin,atan, simplify, eye, solve, latex, atan2, pprint, init_printing, Derivative, sqrt, Eq, solve
from sympy.physics.mechanics import ReferenceFrame, dynamicsymbols, init_vprinting, Point
from sklearn.metrics import mean_squared_error
from scipy.integrate import cumulative_trapezoid
```

```python
#Marco de referencia
N = ReferenceFrame("N")
v = symbols("v") #velocidad en X NO cambia en el tiempo
g = 9.81
t = symbols("t")
posicion = (v)*t*N.x - (0.5*g*t**2)*N.y
#hallar el tiempo
y_conocido = -(1.2+0.9)
t_ecuacionB = Eq(posicion.dot(N.y), y_conocido)
t_B = solve(t_ecuacionB,t)[1] #tomar solucion positiva
#hallar la velocidad
velocidad_X = 2.4/t_B
print(t_B)

#calcular la d
y_caida = -(1.2+0.9+3)
t_caida =  Eq(posicion.dot(N.y), y_caida)
t_B = solve(t_caida,t)[1]

print("tiempo caida", t_B)

d = velocidad_X * t_B - 2.4


tiempos = np.linspace(0,10,100)
posiciones = [posicion.subs({v:velocidad_X, t:tiempo}).dot(N.x) for tiempo in tiempos]
plt.plot(tiempos, posiciones)

print("Magnitud velocidad", velocidad_X)
print("d",d)




```
