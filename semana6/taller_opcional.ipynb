```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sympy import diff,symbols,dsolve, Matrix, pi, cos, sin, simplify, eye, solve, latex, atan2, pprint, init_printing, Derivative, sqrt, integrate, Eq
from sympy.physics.mechanics import ReferenceFrame, dynamicsymbols, init_vprinting

# Configuración de impresión en formato LaTeX (MathJax)
init_vprinting(use_latex='mathjax')
```

PUNTO 1


```python
t = symbols("t")
r, theta = dynamicsymbols("r theta")

N = ReferenceFrame("N")
A= N.orientnew("A", "Axis", (theta, N.z))
r_vector = r * A.x

derivada_vector = r_vector.diff(t,N)

derivada_doble_vector = derivada_vector.diff(t,N)

display(r_vector.express(A).simplify())
display(derivada_vector.express(A).simplify())
display(derivada_doble_vector.express(A).simplify())

#La aceleracion radial es la que sigue a nuestro vector
#Es la encargada de darle magnitud 
#Por tanto, queda en la componente x
acel_radial = derivada_doble_vector.express(A).simplify().dot(A.x)
display(acel_radial)

#La aceleracion tangencial es la que le da sentido a la aceleracion
acel_tangencial = derivada_doble_vector.express(A).simplify().dot(A.y)
display(acel_tangencial)

```

PUNTO 2

```python
t, b, aceleracion_tangencial_constante = symbols("t b a_tan_cte")
theta = dynamicsymbols("theta")

O = ReferenceFrame("O")
A = O.orientnew("A", "Axis", (theta, O.z))

vector_r = b*A.x
vel_r = vector_r.diff(t,O)
acel_r = vel_r.diff(t,O)

display(acel_r.express(A).simplify())

ecuacion = Eq(diff(theta, t, 2), aceleracion_tangencial_constante)
theta_solo = dsolve(ecuacion ,ics={theta.subs(t,0): 0})
theta_dot = diff(theta_solo.rhs, t)
display(theta_solo)
display(theta_dot)

#Como es en la primera revolucion, theta vale en 0 vale 0
```
