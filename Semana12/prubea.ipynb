PUNTO 1

```python
#Ecuaciones para las partes
N = ReferenceFrame("N")
theta, l = dynamicsymbols("theta, l")
A = N.orientnew("A", "Axis", (theta, N.z))

#Contantes
l0, g, t, m_1, m_2, k, l_resorte = symbols("l0,g,t ,m_1, m_2 k, l_resorte")
#incognitas
F_01_x, F_01_y, F_N, M_N, M_motor, F_actuador = symbols(" F_01_x, F_01_y, F_N, M_N, M_motor, F_actuador")  

#Vuelto vector
F_01 = F_01_x*N.x + F_01_y*N.y

#Vector del resorte
# Definimos los puntos A y B
x1, y1, x2, y2 = symbols("x1, y1, x2, y2")
P = x1*N.x + y1*N.y
G = x2*N.x + y2*N.y

# Vector de P hacia G
GP = G - P

#Inercias
I_cm_1, I_cm_2 = symbols("I_cm_1 I_cm_2")

#Posición cuerpo 1 centro de masa
p_1 = (l0/2)*A.x
a_1= p_1.diff(t,N).diff(t,N)

#Ecuaciones de Newton-Euler cuerpo 1
sumF_1 = F_01 + F_N*A.y - m_1*g*N.y - m_1*a_1
sumM_1 = M_motor*N.z + M_N*N.z + (p_1.cross(F_N*A.y)) + (-1*p_1.cross(F_01)) - I_cm_1*theta.diff(t,t)*N.z


#P Posición cuerpo 2 centro de masa
p_2 = (l0 + l/2)*A.x
a_2 = p_2.diff(t,N).diff(t,N)

#Fuerza resorte
F_resorte = k*l_resorte*(GP)

#Pos fuerza N
pos_f_N = -(l/2)*A.x
pos_f_resorte = (l/2)*A.x

#Ecuaciones de Newton-Euler cuerpo 2
sumF_2 = F_actuador*A.x - F_N*A.y - m_2*g*N.y - m_2*a_2
sumM_2 = - M_N*N.z + (pos_f_N.cross(-1*F_N*A.y)) + (pos_f_resorte.cross(F_resorte)) - I_cm_2*theta.diff(t,t)*N.z

#Tomar ecuaciones
eqs = [sumF_1.dot(N.x), sumF_1.dot(N.y), sumF_2.dot(N.x), sumF_2.dot(N.y), sumM_1.dot(N.z), sumM_2.dot(N.z)] 

AA , b = sympy.linear_eq_to_matrix(eqs, [F_01_x, F_01_y, F_N, M_N, M_motor, F_actuador])

Motor_value = AA.solve(b)[4].simplify()
F_actuador_value = AA.solve(b)[5].simplify()
```

```python
#Ecuaciones para las partes
N = ReferenceFrame("N")

#Contantes
m_polea, m, t ,g, k,delta_x, R = symbols("m_polea, m, t g, k, delta_x , R")

#incognitas
t_y, alpha = symbols("t_y, alpha")

#Fuerza resorte
F_resorte = k*delta_x*N.y
#Tensión
T = t_y*N.y
#Borde polea
d = R*N.x
#Inercia de la polea
I_polea = (1/2)*m_polea*R**2
#Ecuaciones Newton Euler cuerpo 1 (polea)
sumF_1 = -m_polea*g*N.y - F_resorte - T #la polea no se desplaza
sumM_1 = d.cross(T) -d.cross(F_resorte) - I_polea*alpha*N.z
#Ecuaciones Newton Euler cuerpo 2(masa)
sumF_2 = -m*g*N.y + T - m*(alpha*R)*N.y

eqs = [sumF_1.dot(N.y), sumF_2.dot(N.y), sumM_1.dot(N.z)] 

AA , b = sympy.linear_eq_to_matrix(eqs, [t_y, alpha])

AA.solve(b)
```
